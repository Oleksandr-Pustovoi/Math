Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def converterDtoP(x,y):
    """возвращает полярные координаты R и α в радианах из декартовых"""
    R = (x**2+y**2)**.5
    return R,math.acos(x/R)
```

```python
a=0
b=0
R=10
r = []
a1 = []
a2 = []
p = 5
for i in range(-1000, 1001):
    x_ = i/100
    y1_ = (R**2-(x_-a)**2)**.5+b
    r_O, a1_O = converterDtoP(x_,y1_)
    r.append(r_O)
    a1.append(a1_O)
    
    y2_ = ((R**2-(x_-a)**2)**.5+b)
    r_O, a2_O = converterDtoP(x_,y2_)
    a2.append(a2_O+np.pi)
    
plt.axes(polar = True)
plt.plot(a1,r)
plt.plot(a2,r)
plt.title("график окружности в полярных координатах", va='bottom')
plt.show()
```

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.
